```python
from sklearn.preprocessing import LabelEncoder
data = merged_df.copy()
le = LabelEncoder()
feature = ["Platform", "Genre"]
for col in feature:
    data[col] = le.fit_transform(merged_df[col])
```

```python
data.head()
```

```python
#Numerical values are assigned to categorical data with labelencoder
```
